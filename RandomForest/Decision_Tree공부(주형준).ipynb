##Decision Tree

테니스 표.jpg

테니스 경기가 치뤄지는지의 여부에 대해 다음과 같은 데이터가 존재한다. 이 기록을 가지고 다음에 테니스 경기가 치뤄질 수 있을지에 대해 예측해보려한다.

기본적인 아이디어는 테니스 경기 유무와 가장 비슷한 다른 변수를 찾아 내는 것이다.

테니스 표 2.jpg

위의 표에서 변수 A,B,C 중 테니스 경기 유무와 가장 높은 유사도를 보이는 지표는 B일 것이다.

그렇다면 A와 C중 어떤 것이 더 유사도가 높을까?

이런 질문에 대답하기 위해서 우리는 엔트로피라는 지표를 사용한다.

엔트로피.jpg

엔트로피는 불순도를 나타내는 개념이며 값이 클수록 더 높은 불순도를 나타낸다. 따라서 지금의 경우 엔트로피가 낮은 값을 구하여야 한다.

이를 위의 예시에 적용하여 보자.

E(Play| A) = $ \frac{2}{5}(- 1log_2 1 - 0 log_2 0) + \frac{3}{5}(- \frac{1}{3}log_2 \frac{1}{3} -  \frac{2}{3} log_2  \frac{2}{3}) = 0.55$

E(Play| B) = $ \frac{2}{5}(- 1log_2 1 - 0 log_2 0) + \frac{3}{5}(- 1log_2 1 - 0 log_2 0) = 0$

E(Play| C) = $ \frac{2}{5}(- \frac{1}{2}log_2 \frac{1}{2} -  \frac{1}{2} log_2  \frac{1}{2}) + \frac{3}{5}(- \frac{1}{3}log_2 \frac{1}{3} -  \frac{2}{3} log_2  \frac{2}{3}) = 0.95$

따라서 우리는 속성 B가 play속성을 가장 잘 반영하고 그 다음은 A 그 다음은 C라는 것을 수치로 구할 수 있다.

그렇다면 다시 테니스 예시로 돌아가서 어떤 속성이 play를 가장 잘 반영하는지 구해보자.

먼저 E(Play)를 구해준다.

E(Play) = $ -\frac{9}{14} log_2 \frac{9}{14} - \frac{5}{14} log_2 \frac{5}{14} = 0.940$

그리고 각각의 요소들에 대하여 엔트로피 값을 구하면

날씨.jpg

E(Outlook) = $ \frac{5}{14}(-\frac{2}{5} log_2 \frac{2}{5} - \frac{3}{5} log_2 \frac{3}{5}) + $
$ \frac{4}{14}(-\frac{4}{4} log_2 \frac{4}{4} - \frac{0}{4} log_2 \frac{0}{4}) + $
$ \frac{5}{14}(-\frac{3}{5} log_2 \frac{3}{5} - \frac{2}{5} log_2 \frac{2}{5}) = 0.694$

Gain(Play,Outlook) = E(Play) - E(Outlook) = $0.940 - 0.694 = 0.246$

온도.jpg

E(Temp) = $ \frac{4}{14}(-\frac{2}{4} log_2 \frac{2}{4} - \frac{2}{4} log_2 \frac{2}{4}) + $
$ \frac{6}{14}(-\frac{4}{6} log_2 \frac{4}{6} - \frac{2}{6} log_2 \frac{2}{6}) + $
$ \frac{4}{14}(-\frac{3}{4} log_2 \frac{3}{4} - \frac{1}{4} log_2 \frac{1}{4}) = 0.911$

Gain(Play,Temp) = E(Play) - E(Temp) = $0.940 - 0.911 = 0.029$

습도.jpg

E(Humidity) = $ \frac{7}{14}(-\frac{3}{7} log_2 \frac{3}{7} - \frac{4}{7} log_2 \frac{4}{7}) + $
$ \frac{7}{14}(-\frac{6}{7} log_2 \frac{6}{7} - \frac{1}{7} log_2 \frac{1}{7}) = 0.798$

Gain(Play,Humidity) = E(Play) - E(Humidity) = $0.940 - 0.798 = 0.151$

바람.jpg

E(Wind) = $ \frac{6}{14}(-\frac{3}{6} log_2 \frac{3}{6} - \frac{3}{6} log_2 \frac{3}{6}) + $
$ \frac{8}{14}(-\frac{6}{8} log_2 \frac{6}{8} - \frac{2}{8} log_2 \frac{2}{8}) = 0.892$

Gain(Play,Wind) = E(Play) - E(Wind) = $0.940 - 0.892 = 0.048$

이때 Outlook의 gain이 0.246으로 가장 크기 때문에 우리의 decision tree는 Outlook을 root node로 가지게 된다.

트리초반.jpg

이때 3가지 상황에서도 마찬가지로 Temp, Humidity, Wind의 엔트로피를 고려하여 가장 유사도가 높은 요소를 선택하여 나머지 트리를 완성하면 된다.

트리마무리.jpg

안성된 트리의 모습은 다음과 같으며 이 트리를 기반으로 우리는 다음 테니스 경기가 치뤄질지에 대해 예측할 수 있다. 예를 들어 어떤 날이 Rain, Mild, Normal, Weak라면 트리에서 가장 오른쪽 child node로 향해 "테니스 경기가 치뤄질 것이다" 라고 예측할 것이다.
